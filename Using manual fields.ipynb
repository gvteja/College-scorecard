```python
import pandas as pd
pd.set_option('display.height', 1000)
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
import numpy as np
```

```python
df=pd.read_csv('./git/MERGED2011_PP.csv')
```

```python
dd=pd.read_csv('./git/CollegeScorecardDataDictionary-09-12-2015.csv')
```

```python
target_cols=['mn_earn_wne_p10',
'md_earn_wne_p10']
```

```python
#if the target column is 'PrivacySuppressed' then we can them use them. So dropping them.
print 'Original shape of data :',df.shape
df=df.loc[(df['mn_earn_wne_p10']!='PrivacySuppressed') &(df['md_earn_wne_p10']!='PrivacySuppressed')]
print 'New shape :',df.shape
df.to_csv('./git/MERGED2011_PP_cleaned.csv')
df=pd.read_csv('./git/MERGED2011_PP_cleaned.csv')
```

```python
df.mn_earn_wne_p10.describe()
```

```python
df.md_earn_wne_p10.describe()
```

```python
treasury_dd=dd[dd['SOURCE']=='Treasury']
print 'Total number of fields from treasury :',len(treasury_dd['VARIABLE NAME'])
frames_from_treasury=df[treasury_dd['VARIABLE NAME']]
frames_from_treasury.describe()
```

```python
ipeds_dd=dd[dd['SOURCE']=='IPEDS']
all=ipeds_dd['VARIABLE NAME']
all=all[1:]
print 'Total number of fields from IPEDS ',len(all)
frames_from_ipeds=df[all]
descr=frames_from_ipeds.describe()
frames_from_ipeds.describe()
```

```python
good_fields_from_ipeds=[]
for d in descr:
    if descr[d]['count']>3500:
        good_fields_from_ipeds.append(d)
print 'Goodies fields ipeds :',len(good_fields_from_ipeds)
```

```python
NSLDS_dd=dd[dd['SOURCE']=='NSLDS']
all=NSLDS_dd['VARIABLE NAME']
print 'Total number of fields from NSLDS :',len(all)
frames_from_NSLDS=df[all]
descr=frames_from_NSLDS.describe()
frames_from_NSLDS.describe()
```

```python
good_fields_from_nslds=[]
for d in descr:
    if descr[d]['count']>4000:
        good_fields_from_nslds.append(d)
print 'Goodies nslds :',len(good_fields_from_nslds)
```

```python
good_fields=good_fields_from_ipeds
```

```python
all_fields=good_fields+['mn_earn_wne_p10','md_earn_wne_p10']
```

```python
all_fields_df=df[all_fields]
print 'Shape :',all_fields_df.shape
all_fields_df.describe()
```

```python
all_fields_df.dropna(inplace=True)
print 'Shape after dropna() :',all_fields_df.shape
```

```python
X=all_fields_df[good_fields]
Y1=all_fields_df['mn_earn_wne_p10']
Y2=all_fields_df['md_earn_wne_p10']
```

```python
from sklearn.cross_validation import train_test_split
X_train,X_Test,Y_train,Y_test=train_test_split(X.as_matrix(),Y1.as_matrix().reshape(-1,),test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
clf=LinearRegression()
clf=clf.fit(X_train,Y_train)
```

```python
from sklearn.metrics import mean_squared_error
print mean_squared_error(Y_test,clf.predict(X_Test))
```
